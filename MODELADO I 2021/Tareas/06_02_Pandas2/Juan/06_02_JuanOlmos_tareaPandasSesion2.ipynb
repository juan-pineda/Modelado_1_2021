En el siguiente un dataset de los flujos de fondo de los Estados Unidos:

* Convierta las fechas en un DatetimeIndex
* Cambie la frecuencia a mensual, sume los valores y asígnela ese valor a cada mes
* Eliminar las filas que obtuvieron valores Nan

```python
import pandas as pd
import numpy as np
```

```python
#Cargamos los datos 

df = pd.read_csv('../data/fondo_us.csv')
df.head()
```

## 1.) Fechas a datetime

```python
#antes del datetime
print(df.dtypes)
df.Date = pd.to_datetime(df.Date)
print(df.dtypes)
```

## 2.) Frecuencia indices

```python
#cambiando a indice
df = df.set_index('Date')
df.head()
```

```python
df_months = df.resample('M').last()
df_months.head()
```

## 3.) Eliminar Nans

```python
df_months = df_months.dropna(axis=0, how='any')
df_months
```
