# Loading data

Los datos que se cargan son 176 puntos. Luego de un proceso previo, cada uno de estos puntos representa a un paciente. Los primeros 88 puntos de la lista son pacientes con Parkinson, el resto son pacientes Control (No-Parkinson)

```python
data=np.load('/content/drive/MyDrive/Modelado_2021/Tareas/Tarea03 29/JuanOlmos/TangentSpace.npy')
print(f'Shape data: {data.shape}')
xmin, xmax = min(data[:,0]), max(data[:,0]) #tomando rango eje x
print(f"xmin = {xmin}")
print(f"xmax = {xmax}")

"""
Splitting data
"""
P , NP = data[:88,:] , data[88:,:] #se dividen los puntos en puntos de P y NP
Px , Py = P[:,0], P[:,1] #Se toman las coordenadas de los P
NPx , NPy = NP[:,0] , NP[:,1] #Se toman las coordenadas de los NP
```

# Fig 1 (Baseline)

Se pintan los puntos de P y NP

```python
plt.figure(figsize=(10,10))
plt.title('Patients representation')
plt.scatter(Px,Py, c='r', label='Parkinson (P)')
plt.scatter(NPx, NPy, c='b', label='No Parkinson (NP)')
plt.xlabel('Desviación movimiento')
plt.ylabel('Desviación edad 70 años')
plt.legend()
```

# Fig 2

```python
plt.figure(figsize=(10, 10))
plt.title('Patients representation', fontsize=15)
plt.scatter(Px,Py, c='tab:red',alpha=0.3, s=300, edgecolors='none', linewidth=4 ,label='Parkinson (P)')
plt.scatter(NPx, NPy, color='tab:blue', alpha=0.3 , s=300, edgecolors='none', linewidth=4 ,label='No Parkinson (NP)')
plt.axis('off')
# plt.xlabel('Desviación movimiento')
# plt.ylabel('Desviación edad 70 años')
plt.legend(labelspacing=2)
#plt.legend()
plt.show()
```

# Fig 4 (En Clase)
Basada con las recomendaciones hechas en clase

```python
P_xmean, P_ymean, NP_xmean, NP_ymean = [np.mean(i) for i in [Px, Py, NPx, NPy]]

#Para utilizar un font específico
font = {'family': 'serif',
        'weight': 'normal',
        'size': 14,
        }

font_axis = {'family': 'serif',
        'color': 'black',
        'weight': 'normal',
        'size': 14,
        }


plt.figure(figsize=(10, 10))
#plt.title('Patients representation', fontsize=15)
plt.scatter(Px,Py, c='tab:red',alpha=0.3, s=300, edgecolors='none', linewidth=4 ,label='Parkinson (P)')
plt.scatter(P_xmean, P_ymean,c='tab:red', s=300, marker='x', edgecolors='none', linewidth=4 )
plt.scatter(NPx, NPy, color='tab:blue', alpha=0.3 , s=300, edgecolors='none', linewidth=4 ,label='No Parkinson (NP)')
plt.scatter(NP_xmean, NP_ymean,c='tab:blue', s=300, marker='x', edgecolors='none', linewidth=4 )
#plt.xticks([]), plt.yticks([])
plt.text(P_xmean+1, P_ymean-0.2, 'Parkinson mean', fontdict=font, color='tab:red')
plt.text(NP_xmean+1, NP_ymean-0.15, 'No Parkinson mean', fontdict=font, color='tab:blue')
plt.xlabel('Desviación movimiento', fontdict=font_axis)
plt.ylabel('Desviación edad 70 años', fontdict=font_axis)
#plt.axis('off')
plt.legend(labelspacing=2)
plt.show()
```
