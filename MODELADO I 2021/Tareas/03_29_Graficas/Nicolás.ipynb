```python
import numpy as np
import matplotlib.pyplot as plt


data = [[ 2, 75,  110, 20,  12],
        [ 4, 89, 136, 37, 22],
        [ 8, 105, 162, 48, 35],
        [ 11, 121, 174, 65, 51],
        [15, 154, 200, 80, 67 ]]

columns = ('Matemáticas', 'Ing.química', 'Ing.Industrial', 'Medicina', 'Filosofía')
rows = ['%d year' % x for x in (5, 10, 15, 20, 25)]

values = np.arange(0, 250, 50)



colors = plt.cm.BuPu(np.linspace(0, 0.5, len(rows))) #agrega el color 
n_rows = len(data)

index = np.arange(len(columns)) 
bar_width = 0.6


y_offset = np.zeros(len(columns)) #inicia un vector de ceros del tamaño de las columnas

# Crea la gráfica de barras
cell_text = []
for row in range(n_rows):
    plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])
    y_offset = y_offset + data[row]
    cell_text.append(['%1.1f' % x  for x in y_offset])
    

colors = colors[::-1] #voltea los colores para que coincidan
cell_text.reverse()
 
plt.axhline(y=485, color ='gray',  linewidth=1, linestyle="-", alpha=0.5)


#agregar la tabla
the_table = plt.table(cellText=cell_text,
                      rowLabels=rows,
                      rowColours=colors,
                      colLabels=columns,
                      loc='bottom')


plt.subplots_adjust(left=0.1, bottom=0.1) #Hacer espacios para la tabla

plt.ylabel("Número acumulado")
plt.xticks([]) #Quita los lables del vector x
plt.title('Graduados por Carrera UIS')

plt.show()
```
