## Resolviendo el primer punto.

Tomando para `Outcome` el valor de $0$ si el diagnostico es negativo y $1$ si el diagnostico es positivo, tenemos:

```python
diagnostico = db.groupby('Outcome').Glucose.agg(['mean','std'])
diagnostico
```

## Resolviendo el segundo punto. 

Usaremos dos métodos para poder clasificar por grupos de edades.

La primera será usando una función auxiliar llamada `rango_edad()` que permite agregar una columna clasificando la edad de cada paciente. 

La segunda será usando la función de numpy `np.digitize()`a la lista **edad**.

```python
def rango_edad(x):
    if x<=10:
        etapa=1
    elif 10<x<=20:
        etapa = 2
    elif 20<x<=30:
        etapa= 3
    elif 30<x<=40:
        etapa = 4
    elif 40<x<=50:
        etapa = 5
    elif 50<x<=60:
        etapa= 6
    elif 60<x<=70:
        etapa = 7
    elif 70<x<=80:
        etapa = 8
    else:
        etapa = 9
        
    return etapa 
```

```python
db['Rango_edad'] = db.Age.apply(rango_edad)
db.head()
```

```python
glucosa_rango = db.groupby('Rango_edad').Glucose.agg(['mean','std'])
glucosa_rango
```

## Segunda forma para clasificar la edad

```python
edad = db.Age.values
bins = [0.0, 10.0, 20.0, 30.0, 40.0, 50.0, 60.0, 70.0, 80.0, 90.0]
db['Rango_edad2'] = np.digitize(edad,bins,right = True)
db.head()
```
