Creamos la función `rango_edades` para agrupar por edades y la aplicamos a la columna 'Age'.

```python
#df['Age'].unique()
```

```python
def rango_edades(x):
    if x<=10:
        rango='rango_1'
    elif 10<x<=20:
        rango = 'rango_2'
    elif 20<x<=30:
        rango= 'rango_3'
    elif 30<x<=40:
        rango = 'rango_4'
    elif 40<x<=50:
        rango = 'rango_5'
    elif 50<x<=60:
        rango= 'rango_6'
    elif 60<x<=70:
        rango = 'rango_7'
    elif 70<x<=80:
        rango = 'rango_8'
    else:
        rango = 'rango_9'
        
    return rango   
```

```python
df['RangoEdades'] = df.Age.apply(rango_edades) #aplicar la función rango_edades a la columna 'Age' 
df
```

```python
df.groupby('RangoEdades').agg(['mean','std'])
```
