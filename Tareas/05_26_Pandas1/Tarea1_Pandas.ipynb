Calcular:

* la media y la desviación estandar de la glucosa de los pacientes que fueron diagnosticasdos con diabétes y de los que fueron diagnosticados como no-diabéticos

*  Posteriormente, quiere dividir a la población en grupos por rangos de edades de 10 años, es decir, de 0-10, de 10-20, etc. Y para cada uno de estos grupos, calcular la media y la desviación estandar de la glucosa


```python
import pandas as pd
```

```python
datos = pd.read_csv('Data/diabetes.csv')
datos
```

```python
datos_1 = datos.groupby('Outcome').Glucose.agg(['mean','std'])
datos_1.head()
```

```python
def rangos(x):
    if x<=10:
        edad=10
    elif 10<x<=20:
        edad = 20
    elif 20<x<=30:
        edad= 30
    elif 30<x<=40:
        edad = 40
    elif 40<x<=50:
        edad = 50
    elif 50<x<=60:
        edad= 60
    elif 60<x<=70:
        edad = 70
    elif 70<x<=80:
        edad = 80
    else:
        edad = 90
        
    return edad
```

```python
datos['Rangos'] = datos.Age.apply(rangos)
datos
```

```python
datos.groupby('Rangos').Glucose.agg(['mean','std'])
```
