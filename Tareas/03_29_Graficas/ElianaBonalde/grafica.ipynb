### Teorema del límite central

Este teorema describe la distribución de medias de muestras aleatorias provenientes de una población con media y con varianza no nula y finita, las cuales se distribuyen normalmente. 


```python
import matplotlib.pyplot as plt
import numpy as np
#import matplotlib.mlab as mlab
import scipy.stats as stats
```

```python
mu, sigma = 0, 0.2 #Media, desviación estandar
Xmed =  np.zeros(1000) #Vector de ceros con 1000 elementos

for j in range(0,999): 
    datos = np.random.normal(mu,sigma,1000) #Crea un vector aleatorio de 1000 datos distribuidas normalmente
    Xmed[j] = np.mean(datos) #Los ceros los sustituye con las medias de todos los vectores generados en datos

    
plt.subplot(2, 1, 1)    
plt.hist(Xmed, 50) 
n, bins, patches = plt.hist(Xmed, 50) #Saca los valores (n, bins y patches) del histograma
plt.ylabel('Histograma')
plt.title('Teorema del Límite Central')
y = stats.norm.pdf(bins, mu, sigma/(1000**(0.5)))
#y = mlab.normpdf(bins, mu, sigma/(1000**(0.5)))
plt.plot(bins, y, '--', label = '$\mu=0$')


plt.subplot(2, 1, 2)
plt.boxplot(Xmed, 0, 'b+', 0) 
plt.ylabel('Gráfica de caja')


plt.show()
```
