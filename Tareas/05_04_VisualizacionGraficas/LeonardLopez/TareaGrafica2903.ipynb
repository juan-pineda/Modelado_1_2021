Estudiando la ecuación de onda en EDPS
$$c²\frac{\partial²u }{\partial x²}=\frac{\partial²u }{\partial t²}$$
Bajo condiciones dadas es posible deducir las **ondas estacionarias** o **modos normales**
como solución de un problema de valores en la frontera. $c$ es la velocidad de vibración y depende de la tensión $T$ y de la densidad de masa por unidad de longitud $\rho$. 
$$c=\sqrt{\frac{T}{\rho}}$$
Cuando $T$ es suficientemente grande, la cuerda vibrando produce un sonido musical. Este sonido es el resultado de ondas estacionarias.

Sus soluciones son de la forma:
$$u_n(x,t)=C_n\sin\left(\frac{n\pi c}{L}t+\phi_n\right) \sin\left(\frac{n\pi }{L}x\right)$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
L=5
n=4
Fi=0.2#ángulo de fase temporal
c=10 #velocidad de la onda
A=0.5#amplitud máxima, relacionada con la intensidad del sonido
t=0.18 #instantánea en el tiempo t segundos
x=np.arange(0.0,L,0.01)
u=A*np.sin(n*np.pi*c*t/L+Fi)*np.sin(n*np.pi*x/L)#aquí definimos la función 

#plotting los datos
plt.xlabel('Eje x')
plt.ylabel(f'u (x,{t})')
plt.title(f'{n}-ésima onda estacionaria')
plt.grid()#poner grilla
plt.plot(x,u)
```

## Solitón u onda viajera.

Uno de los ejemplos más clásicos en la literatura es el poder poder estudiar la ecuación de Onda si $-\infty <x< \infty$, con $t>0$. 
$$c²\frac{\partial²u }{\partial x²}=\frac{\partial²u }{\partial t²},$$
$u(x,0)=f(x)$, $\dfrac{\partial u}{\partial t}(0,x)=g(x)$. Tiene por solución la función
$$u(x,t)=\frac{1}{2}[f(x+ct)+f(x-ct)]+\frac{1}{2c}\int_{x-ct}^{x+ct}g(s)ds.$$

Mostramos ahora la solución $g(x)=0$, cuya solución se puede interpretar como una superposición de dos **ondas viajeras**, una moviéndose hacía la derecha y otra hacía la izquierda. Cabe resaltar, que ambas ondas viajan a la misma raídez y tienen la misma forma. 

A la anterior solución se le conoce en la literatura como **solución de d'Alembert**.

Consideremos $f(x)= 1-|x|$ si $|x|\leq 1$ y $f(x)=0$ en todos los demás puntos en la recta real.

```python
plt.figure(figsize=(14,10))
ax=plt.subplot(1, 1, 1) 
c=1
L=6
x=np.arange(-L,L,0.01)
def f(x):
    f=np.piecewise(x,[x<-1,np.abs(x)<=1,x>1],[lambda x:0,lambda x: 1-np.abs(x),lambda x:0])
    return f
plt.grid()#poner grilla
for k in range(10):
    t=0.4*k
    if k==0:
        plt.plot(x,0.5*(f(x+c*t)+f(x-c*t)),color='r',linewidth=3,label=f'Solitón t={k}')
    else:
        plt.plot(x,0.5*(f(x+c*t)+f(x-c*t)))

ax.spines['right'].set_color('none')    
ax.spines['top'].set_color('none')
plt.xlabel('Eje x')
plt.ylabel('u (x,t)')
plt.title('Onda viajera ')
plt.legend(frameon=False)
plt.show()
    

```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('seaborn-pastel')
L=6
n=4
Fi=0.2#ángulo de fase temporal
c=10 #velocidad de la onda
A=0.5#amplitud máxima, relacionada con la intensidad del sonido
#t=0.18 #instantánea en el tiempo t segundos

fig = plt.figure()
ax = plt.axes(xlim=(-L, L), ylim=(-0.2, 1.1))
line, = ax.plot([], [], lw=3)

def init():
    line.set_data([], [])
    return line,

x=np.arange(-L,L,0.01)
def f(x):
    f=np.piecewise(x,[x<-1,np.abs(x)<=1,x>1],[lambda x:0,lambda x: 1-np.abs(x),lambda x:0])
    return f

def animate(t):
    #x=np.arange(0.0,L+0.01,0.01)
    u=0.5*(f(x+c*0.005*t)+f(x-c*0.005*t))#aquí definimos la función 
    line.set_data(x, u)
    return line,

anim = FuncAnimation(fig, animate, init_func=init,
                               frames=170, interval=20, blit=True)


anim.save('soliton.gif', writer='imagemagick')
```
