```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
weekly = pd.read_csv('https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv')
```

```python
weekly.head()
```

```python
# Guarda datos en CSV:
weekly.to_csv('weekly.csv', header=True, index=True)
```

```python
wl = pd.read_csv('weekly.csv')
wl.head(5)
```

```python
wl.dtypes #date es un object
```

```python
wl.Date = pd.to_datetime(wl.Date)# cambiamos Date de tipo de object a datetime64

wl['Date'].head()
```

```python
wl = wl.set_index('Date') #convertimos Date en un DatetimeIndex

wl.head()
```

```python
wl = wl.sort_index(ascending=True)
wl.head()
```

```python
# Cambiemos la frecuencia a mensual, sumemos los valores y asígnemos ese valor a cada mes
# Se obtiene el último día hábil de cada mes
wl_month = wl.resample('M').last()
wl_month
```

```python
#Veamos si hay elementos NaN usando isna()
wl_check = wl.isna()
wl_check
```

```python
#Veamos si hay elementos NaN usando isna()
wl_check1 = wl.isnull()
wl_check1
```

```python
#Veamos si por lo menos un NaN usando isna()
wl_check2 = wl.isna().any().any()
wl_check2
```

```python
# Eliminando las filas que obtuvieron valores Nan
# haciendo un nuevo marco de datos con valores NaN eliminados
new_wl = wl.dropna(axis = 0, how ='any')
new_wl
```

```python
# comparar tamaños de marcos de datos
print ("Longitud del marco de datos antiguo:", len (wl), "\ nLongitud del marco de datos nuevo:",
        len (new_wl), "\ nNúmero de filas con al menos 1 valor NA:",
        (len (wl) -len (new_wl)))
```

```python
# Guarda datos en CSV:
new_wl.to_csv('new_weekly.csv', header=True, index=True)
```
