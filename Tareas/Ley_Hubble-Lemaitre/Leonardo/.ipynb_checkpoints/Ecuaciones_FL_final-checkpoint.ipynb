## Definición de las constantes $\Omega_M$, $\Omega_L$, $\Omega_K$.

Siendo $\Omega_M$ el porcentaje de *energía del Universo que está compuesto por materia*, $\Omega_L$ representando el porcentaje de *energía del Universo en forma de materia oscura,* también conocida como la **quintaesencia** que proporciona la *gravedad negativa* necesaria para que el universo se expanda; y $\Omega_K = 1- \Omega_M- \Omega_L$, la cual representa la *curvatura del universo*. 

**Obs:** Si $\Omega_K= 0$ quiere decir que el Universo es plano, para casos de mayor o menor que cero (0), se tienen modelos de Universo *abierto/cerrado.*

```python
c = ct.c.to('km/s').value
H_0 =cosmo.H(0).value 
```

tenemos las siguientes definiciones de funciones para nuestro problema:


```python
# devuelve el valor de la integral I para un valor de z dado.
def I(z, Omega_M = 0.3, Omega_L = 0.7):
    return integrate.quad(lambda z: 1/np.sqrt(Omega_M*(1+z)**3+Omega_L+(1 - Omega_M - Omega_L)*(1+z)**2), 0, z)[0]

# Permite calcular la distancia de luminosidad D_L para un valor z
def D_L(z, Omega_M = 0.3, Omega_L = 0.7):
    Omega_K = 1 - Omega_M - Omega_L
    aux_1 = np.sqrt(np.abs(Omega_K))              
    if (Omega_M + Omega_L) > 1:
                    aux = (((1+z)*c)/(H_0*aux_1))*np.sin(aux_1*I(z, Omega_M, Omega_L))
    elif (Omega_M + Omega_L) == 1:
                    aux = (((1+z)*c)/H_0)*I(z, Omega_M, Omega_L)
    else:
                    aux = (((1+z)*c)/(H_0*aux_1))*np.sinh(aux_1*I(z, Omega_M, Omega_L))
    return aux

# calcula la función mu(z) conocido como módulo de distancia, relacionado con D_L
def mu(z, Omega_M = 0.3, Omega_L = 0.7):
    x = np.linspace(0.015, z)
    y = np.zeros(np.shape(x)[0])
    for i in np.arange(np.shape(x)[0]):
        y[i] = 5*math.log10(D_L(x[i],Omega_M, Omega_L))+25
    return (x,y)
```

```python
# Definamos la función a integrar de o a z para calcular I
def g(z):
    return 1/np.sqrt(Omega_M*(1+z)**3+Omega_L+
                     (1 - Omega_M - Omega_L)*(1+z)**2)

# Permite calcular la distancia de luminosidad D_L en función del corrimiento al rojo de la luz, z
def funtionD_L(z):
    x = np.linspace(0.01, z)
    y = np.zeros(np.shape(x)[0])
    for i in np.arange(np.shape(x)[0]):
        y[i] = integrate.quad(g, 0, x[i])[0]
    return (x,y) 

# calcula el valor de la función I(z) para distintos valores de 0 a un z dado
def IntI(z):
    x = np.linspace(0, z)
    y = np.zeros(np.shape(x)[0])
    for i in np.arange(np.shape(x)[0]):
        y[i] = D_L(x[i])
    return (x,y)
```

# Ploteando los datos de las diferentes supernovas y diferentes modelos 

```python
plt.rcParams.update({'font.size': 28})
fig = plt.figure(figsize=(25,15))
ax = fig.add_subplot(111)
ax.scatter(informacion['Redshift'],informacion['Distance Modulus'],c = 'k', alpha =0.75)
ax.errorbar(informacion['Redshift'],informacion['Distance Modulus'],
             yerr= informacion['Error'], linestyle = '',c = 'teal', alpha =0.35)
ax.plot(mu(1.45, 0.3, 0.7)[0], mu(1.45, 0.3, 0.7)[1], label = 'Om=0.3 Ol = 0.7', c = 'r', linewidth=2.0)
ax.plot(mu(1.45, 0.75, 0.3)[0], mu(1.45, 0.75, 0.3)[1], label = 'Om=0.75 Ol = 0.3', c = 'm', linewidth=1.5)
ax.plot(mu(1.45, 0.25, 0.69)[0], mu(1.45, 0.25, 0.7)[1], label = 'Om=0.25 Ol = 0.69', c = 'b', linewidth=1.5)
ax.plot(mu(1.45, 0.35, 0.7)[0], mu(1.45, 0.35, 0.7)[1], label = 'Om=0.35 Ol = 0.7', c = 'g', linewidth=1.5)
ax.set_title("Hubble_slide")
ax.set_ylabel('Módulo de distancia')
ax.set_xlabel('Corrimiento al rojo')
plt.legend(loc = 'lower right', frameon = False)
plt.savefig('Basic.pdf', bbox_inches='tight')
plt.show;
```

# Calculando los errores

```python
z = informacion['Redshift'].tolist()
DM_exp = informacion['Distance Modulus'].tolist()
Error =informacion['Error'].tolist()
#Sumemos todos los errores cometidos en cada punto, mediante el error cuadrático medio

def CalculandoErrores(Omega_M, Omega_L):
    suma = []*1
    for i in np.arange(len(z)):
        DM_teo = 5*math.log10(D_L(z[i],Omega_M, Omega_L))+25
        suma.append((DM_exp[i] - DM_teo)**2/Error[i]**2)
    return sum(suma)/len(z)
#así, por ejemplo, para Omega_M = 0.3 y Omega_L = 0.7 tenemos que:
CalculandoErrores(0.3,0.7)
```

# Creando una columna en el Dataframe para los errores 

Tomando a $\Omega_M = 0.3$ y $\Omega_L= 0.7$, tenemos lo siguiente.

```python
Omega_M = 0.3
Omega_L = 0.7
informacion['Error Modelo 1'] = [((5*math.log10(D_L(informacion['Redshift'][z],Omega_M, Omega_L))+25)
                                  -informacion['Distance Modulus'][z])**2/informacion['Error'][z]**2
                                 for z in informacion.index]
```

```python
# Calculando el error cuadrático medio.
np.sum(informacion['Error Modelo 1'])/len(z)
```
