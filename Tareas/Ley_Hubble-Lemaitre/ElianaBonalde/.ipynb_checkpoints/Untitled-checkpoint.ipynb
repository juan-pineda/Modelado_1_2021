```python
import numpy as np
import sympy as sp
#from sympy.plotting import plot
from scipy.interpolate import approximate_taylor_polynomial
import matplotlib.pyplot as plt
```

```python
x = sp.Symbol('x') 
y = sp.exp(x) * sp.cos(x) 
sp.diff(y,x)
```

```python
sp.series(y, x, x0=0, n=4).removeO()
```

```python
f = lambda x:np.exp(x)*np.cos(x)
sin_taylor = approximate_taylor_polynomial(f, 0, 3, 1)
print(sin_taylor)
```

```python
x = np.linspace(-5.0, 5.0, num=100)
y = np.exp(x) * np.cos(x)
plt.plot(x, y, label="cos(x)e^x")
plt.legend()
plt.show()
```

```python
x = np.linspace(-1.0, 2.0, num=100)
plt.plot(x, np.exp(x)*np.cos(x), label="sin curve")
plt.show()
f = lambda x:np.exp(x)*np.cos(x)
for degree in np.arange(1, 10, step=2):
    sin_taylor = approximate_taylor_polynomial(f, 0, degree, 1, order=degree + 2)
    plt.plot(x, sin_taylor(x), 'g', label=f"degree={degree}")    
    plt.plot(x, f(x), label="f(x)=e^x*cos(x)")
    #plt.grid(axis='both')
    plt.legend(loc='best')
    plt.show()
##plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',
##           borderaxespad=0.0, shadow=True)
##
##plt.tight_layout()
```

```python
for degree in np.arange(1, 10, step=2):
    sin_taylor = approximate_taylor_polynomial(f, 0, degree, 1,
                                               order=degree + 2)
    plt.plot(x, sin_taylor(x), label=f"degree={degree}")    
plt.plot(x, f(x),'k--', label="f(x)=e^x*cos(x)")
plt.grid(axis='x')
plt.legend(loc='best')
plt.axis([-1.0, 2, -4, 4])
plt.show()
```
