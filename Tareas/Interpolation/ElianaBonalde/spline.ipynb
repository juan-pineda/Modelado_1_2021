Las funciones Spline son funciones polinómicas a trozos diferenciables hasta un cierto orden en un intervalo dado. En las spline cúbicas, cada polinomio es de grado 3. 

Las funciones Splines Cúbicas son las más usadas entre las Spline para los problemas de interpolación, debido a que son suficientemente suaves para ajustar datos y no tienen el comportamiento oscilatorio que caracteriza a la interpolación con polinomios de alto grado. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
from scipy.interpolate import CubicSpline
```

Creamos el vector `x`y `y` con los puntos dados.

```python
x = np.linspace(1, 5, 5)
y = np.array([13, 15, 12, 9, 13])
xs = np.linspace(1, 5, 200) #Vector para trazar las curvas
```

**Cubic Spline**

```python
cs = CubicSpline(x, y)
```

```python
cs(3.4) 
```

```python
plt.scatter(x, y, label='datos')
plt.scatter(3.4, 10.44)
plt.plot(xs, cs(xs), label='f(x)')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend(loc='best')
plt.grid()
plt.show()
```

### Fenómeno de Runge

```python
f = lambda x:1 / ( 1 + 25 * x**2)
x = np.linspace(-1.0, 1.0, num=100)
```

```python
plt.plot(x, f(x))
plt.show()
```

```python
def interpolacion_Lagrange(x, y, xi):
    m  = len(x) 
    m2 = len(xi)
    yi = np.zeros(m2,dtype = float)
    for k in range(0, m2): 
        ss = 0
        for i in range(0,m):
            li = 1
            for j in range(0,m):
                if i != j:
                    li = li * (xi[k] - x[j]) / (x[i] - x[j])
            ss = ss + li * y[i]        
        yi[k] = ss
    return xi, yi 
```

```python
x_runge = np.linspace(-1, 1, 400)
f_runge = lambda x: 1 / (1 + 25 * x**2)
fgauss = f_runge(x_runge) 
grado = 20
xx = np.array([-0.4, 0, 1, 2, 3.5])
yy = np.array([4.5696, 0, 0, 0, 6.5625])

for k in range(1,grado):
    x = np.linspace(-1, 1, k+1)
    y = f_runge(x)
    [xi, yi] = interpolacion_Lagrange(x, y, x_runge)
    plt.plot(x_runge, fgauss, 'r-', label='Funcion exacta')
    plt.plot(xi, yi,'b-', label='P(x): Interpolacion')
    #plt.grid(axis='both')
    plt.ylabel('Y')
    plt.xlabel('X')
    plt.legend(loc='best')
    plt.show()     
```
